Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions. Use the code given
below and do the necessary changes. Change the number of bins and the plot view to make
the visualization better.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a random 100000x1 array from an exponential distribution
scale = 1.0
exp_data = np.random.exponential(scale, 100000)
```

```python
# Generate data from uniform and normal distributions for comparison
uniform_data = np.random.rand(100000)
normal_data = np.random.normal(size=100000)
```

```python
# Plot the histograms
plt.figure(figsize=(12, 6))
```

```python
# Exponential distribution
plt.hist(exp_data, density=True, bins=100, histtype="step", color="green", label="Exponential")
```

```python
# Uniform distribution
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
```

```python
# Normal distribution
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal"
```

```python
plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")
```

```python
# Set the axis view to better visualize the distributions
plt.axis([-1, 6, 0, 1.1])
```

```python
# Add legend, titles, and labels
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
```

```python
# Show the plot
plt.show()
```

```python
# Task 3: Plot Z = X^2 + Y^2 in a 3D plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("$Z = X^2 + Y^2$")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

```python
data = sns.load_dataset('penguins').dropna()  # Example dataset
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
data = data[features]
```

```python
# Calculate Pearson and Spearman correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')
```

```python
# Plot heatmaps
plt.figure(figsize=(12, 5))
```

```python
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Pearson Correlation")
```

```python
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Spearman Correlation")
```

```python

import pandas as pd
```

```python
df = pd.read_csv('./Au_nanoparticle_dataset.csv')
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
print(new_df.head(20))
```

```python
stats = new_df.describe().T
stats['mean'] = new_df.mean()
stats['std'] = new_df.std()
stats['25%'] = new_df.quantile(0.25)
stats['50%'] = new_df.quantile(0.5)
stats['75%'] = new_df.quantile(0.75)
print(stats[['mean', 'std', '25%', '50%', '75%']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))

for i, feature in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[feature], bins=20, edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot) # bivariate histogram on upper half
g.map_diag(sns.histplot, kde=True) # histogram and KDE plot on the diagonal
g.map_lower(sns.kdeplot) # bivariate KDE plot on the lower half
plt.show()
```
